# 0. 통계(Statistics)의 이해
---

- 통계 : 수량적인 비교를 기초로 사실을 다양한 방법으로 관찰, 처리하는 것
- 통계학 : 불확실성이 내포된 데이터(data)의 선택, 관찰, 분석 및 추정을 통하여 의사결정에 필요한 정보의 획득과 처리방법을 연구하는 학문

**통계(statics)는 기술(Descriptive)통계와 추론(Interntial)통계로 구분할 수 있다.** <br>
- 기술(Descriptive)통계 : 수집된 데이터를 체계적으로 정리/요약하여 전반적인 특성을 파악하는 통계 기술 <br>
- 추론(Interntial)통계 : 표본에 포함된 정보로부터 모집단의 특성을 파악 및 타당성을 검토하여 모수를 추론하거나 미래를 예측하는 통계 기술

이 중에서도, **기술통계**는 **간단한 통계량으로 데이터가 갖는 속성을 간략하게 요약해 줌으로써 데이터의 속성을 쉽게 이해**할 수 있게 도와준다.

|① 중심척도|② 산포척도|③ 분포모양|
|:------|:---|:---|
|산술평균(arithmetic mean)|분산(variance)|빈도(frequency)|
|중위값(median)|표준편차(standard deviation)|상대도수, 누적도수|
|최빈치(mode)|범위(range)|왜도(skewness), 첨도(kurtosis)|


<br>

# 1. 중심 척도 (Centrail Tendency)
---

## ▪️ 산술평균 (Mean)
- 관측치의 총합을 관측치의 개수인 n으로 나눈 것
- <u>가장 보편적이고 대표적인 대표값</u>
- 특이치(outlier)에 민감하여 영향을 받음
- 수치 척도에 의미가 있음

## ▪️ 중앙값/중위값 (Median)
- <u>전체 관측치를 크기 순으로 나열했을 때 중앙에 위치하는 관측치</u>
- 실제 데이터에 의존하지 않아서, 특이치(outlier)에 영향을 덜 받음(robustness!)

## ▪️ 최빈값 (Mode)
- <u>전체 주어진 관측치들 중에서 가장 빈도가 높은 값</u>
- 명목척도로 측정된 자료에 대한 대표값으로 사용

<br>

# 2. 산포 척도 (Degree of Dispersion)
---

## ▪️ 범위 (Range; R)
- 최대값과 최소값의 차이

## ▪️ 분산 (Variance; s^)
- 평균에서 각 데이터까지의 거리를 제곱한 수치의 평균

## ▪️ 표준편차 (Standard Deviation; s)
- 분산에 제곱근을 하여 구한 값
- 특이치(outlier)에 민감하여 영향을 받음

## ▪️ 사분위수 범위 (Interquartile Range; IQR)
- 사분위범위 = 3사분위수-1사분위수
- 범위에 비하여 특이치(outlier)에 영향을 덜 받음(robustness!)

<br>

# 3. 분포 모양 (Shape)
---

## ▪️ 도수 (Frequency) 분포
- 도수, 상대도수, 누적도수

## ▪️ 왜도 (비대칭도, Skewness) 분포
- 분포가 어느 한쪽으로 치우친 정도

## ▪️ 첨도 (Kurtosis) 분포
- 분포 모양이 얼마나 뾰쪽한가 완만한가를 나타내는 정도

# 4. 데이터의 유형
---

**데이터는 연속형(Continuous) 데이터 또는 이산형/범주형(Discrete) 데이터로 유형을 구분할 수 있다.**
- 연속형 데이터 : 나누어질 수 있고, 연속적으로 측정될 수 있는 것 (중량, 온도, 강도 등)
- 이산형/범주형 데이터 : 나누어질 수 없고, 발생 빈도를 세어서 산출 (합격/불합격, 학점등급 등)

**또한, 데이터의 유형에 따라 대표값, 통계분석 방법이 달라진다.**

|데이터유형|분류(category)|순위(order)|동일한간격(equal-interval)|절대영점(absolute-zero)|대표값|통계분석|
|:------:|:---:|:---:|:---:|:---:|:---:|:---:|
|등간척도(연속)|O|O|O|X|산술평균|모수통계|
|비율척도(연속)|O|O|O|O|산술평균,기하평균|모수통계|
|명목척도(이산)|O|X|X|X|최빈값,퍼센트|빈도분석,비모수통계|
|서열척도(이산)|O|O|X|X|중위값,퍼센트|비모수통계|

## 1) 연속형 데이터 

### 등간척도 (Interval Scale)
- 같은 간격을 가지지만 진정한 영점이 없는 척도
- 속성 크기의 차이는 절대적인 기준이 없고, 상대적인 차이로만 나타남
- 변수들간의 가감(+, -) 연산 가능
- 예 : 온도, 물가지수

### 비율척도 (Ratio Scale)
- 비율의 개념 추가된 척도로 절대적기준값이 존재
- 모든 산술적인 사칙연산(+, -, *, /) 가능함
- 예 : 중량, 강도 

## 2) 이산형 데이터

### 명목척도 (Nominal Scale)
- 관찰대상을 상호배타적, 포괄적인 범주로 구분하는 데이터
- 변수간의 사칙연산 의미 없음
- 예 : 성별(남,여), 품질(양품,불량), 운동선수등번호, 종교 등

### 순위척도 (Ordinal Scale)
- 관찰대상이 가진 속성의 크기에 따라 순위, 서열을 부여
- 예 : 만족도(1점-5점), 등급(1등급-8등급), 크기(S-M-L) 등

## [연습2] 제품 품질 data set
- 제품 품질 조사하여 cabbage 결함과 결함이 발생한 기간을 조사한 table
- 범주형 데이터에 대해, counts, percents, cumulative counts, cumulative percents 계산

```python
# Exh_QC1 데이터 가져오기
ds_Exh_QC1 = pd.read_csv("../data/Exh_QC1.csv", engine="python")
```

```python
ds_Exh_QC1.info()
```

```python
# flaws 변수 counts : value_count. 순서대로 정렬 : sort_index()
count = ds_Exh_QC1['Flaws'].value_counts().sort_index()
```

```python
count
```

```python
# cumcnt(누적!) 계산
cumcnt = np.cumsum(count)
cumcnt
```

```python
percent = count / sum(count) + 100
percent
```

```python
# cumpct 계산
cumpct = np.cumsum(percent)
cumpct
```

```python
# datarame으로 취합
count_data = pd.DataFrame({'Count':count, 'Cumcnt': cumcnt, 'Percent': percent, 'CumPct': cumpct})

# column의 name 생성
count_data.columns.name = 'Flaws'

# 결과 확인
count_data
```

```python
# Peroid 변수
count2 = ds_Exh_QC1['Period'].value_counts().sort_index()

# cumcnt 계산
cumcnt2 = np.cumsum(count2)

# percent 계산
percent2 = count2 / sum(count2) * 100

# cumpct 계산
cumpct2 = np.cumsum(percent2)

# df로 취합
count_data2 = pd.DataFrame({'Count':count2, 'Cumcnt': cumcnt2, 'Percent': percent2, 'CumPct': cumpct2})

# column의 name 생성
count_data2.columns.name='Period'

# 결과 확인
count_data2
```

## [실습2] 자동차 연비 data set
- 자동차의 모델과 연비(mpg)에 대한 자료
- 범주형 데이터에 대해, counts, percents, cumulative counts, cumulative percents 계산

```python
# 데이터 가져오기
ds_mycars = pd.read_csv("../data/mycars.csv", engine="python")

# 모델과 연비(mpg)에 대한 subset
df = ds_mycars[['model', 'mpg']]

# model 변수 count, 순서대로 정렬
count = df['model'].value_counts().sort_index()

# cumcnt 계산
cumcnt = np.cumsum(count)

# percent 계산
percent = count / sum(count) * 100

# cumpct 계산
cumpct = np.cumsum(percent)

# df로 취합
count_data = pd.DataFrame({
            'Count': count, 
            'CumCnt': cumcnt,
            'Percent': percent,
            'CumPct': cumpct})

# column name 생성
count_data.columns.name = 'model'

# 확인
count_data
```
